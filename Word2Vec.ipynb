```python
# YT video id
video_id = "lacFcgcHx6I"

from youtube_transcript_api import YouTubeTranscriptApi , TranscriptsDisabled

try:
  yt_api = YouTubeTranscriptApi()
  transcript_list = yt_api.fetch(video_id, languages=["en"])
except TranscriptsDisabled:
  print("Transcript disabled for this video")

print(transcript_list[:10])
```

```python
transcript = " ".join([val.text for val in transcript_list])
print(transcript[:100])
```

```python
!pip install gensim
!pip install nltk
```

```python
# we got the text now preprocessing
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt_tab')
from gensim.utils import simple_preprocess
```

```python
tokenise  = sent_tokenize(transcript)
print(tokenise[:5])  # first 5
```

```python
# basic preprocessing
preprocessed_text = [simple_preprocess(sentence) for sentence in tokenise]
print(preprocessed_text[:5])

# another way of writing

# alternative_preprocessed_text = []
# for sentence in tokenise:
#     processed_sentence = simple_preprocess(sentence)
#     alternative_preprocessed_text.append(processed_sentence)

# print(alternative_preprocessed_text[:5])
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

new_text_sentences = []
for sentence_word_list in preprocessed_text:
    filtered_sentence = [word for word in sentence_word_list if word not in stopwords.words('english')]
    if filtered_sentence: # adding non empty sentense
        new_text_sentences.append(filtered_sentence)


print(new_text_sentences[:5])
```

```python
import gensim
model = gensim.models.Word2Vec(
    window = 10, # taking 10-10 words both side meaning input of 20 words
    min_count = 2
)
```

```python
model.build_vocab(new_text_sentences) # word2vec wants  list not just words ['words']->fine but 'words'->not fine
```

```python
model.train(new_text_sentences, total_examples=model.corpus_count, epochs=model.epochs)
```
