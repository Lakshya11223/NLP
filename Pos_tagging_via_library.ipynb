```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp("I will google about facebbok")
```

```python
doc.text
```

```python
doc[-1]
```

```python
doc[2].pos_
```

```python
doc[2].tag_
```

```python
spacy.explain('VB')
```

```python
for word in doc:
  print(word.text,"---->",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc2 = nlp("You read the books on history")

for word in doc2:
  print(word.text,"---->",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
from spacy import displacy
```

```python
displacy.render(doc2,style='dep')
```
